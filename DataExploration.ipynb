```python
#shows data types of all values
census.dtypes
```

```python
#Unique values in classwkrd variable
census["classwkrd"].unique()
```

```python
#mean for inctot for diff educ levels
census.groupby("educ")["inctot"].mean()
```

```python
#show std for inctot
census.groupby("educ")["inctot"].std()
```

```python
#create histo of incwage
import matplotlib.pyplot as plt

plt.hist(census["incwage"], bins=30)
plt.xlabel("incwage")
plt.ylabel("Frequency")
plt.title("Histogram of incwage")
plt.show()
```

```python
#create log wage
import numpy as np

census["log_wage"] = np.log(census["incwage"].replace(0, np.nan))
```

```python
#Histo of log wage
plt.hist(census["log_wage"].dropna(), bins=30)
plt.xlabel("log_wage")
plt.ylabel("Frequency")
plt.title("Histogram of log_wage")
plt.show()
```

```python
#plot scatter graph
import matplotlib.pyplot as plt

plt.scatter(census["uhrswork"], census["log_wage"])
plt.xlabel("Usual Hours Worked per Week (uhrswork)")
plt.ylabel("Log Wage (log_wage)")
plt.title("Relationship Between uhrswork and log_wage")
plt.show()
```
